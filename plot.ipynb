```python
%matplotlib inline
from numpy import exp,sqrt
from plot import plot

def fun(x):
    return 0.5*(-1-sqrt(1+0.2*x)) 

def funs(xs):
    return [fun(x) for x in xs]
plot(0,4000,[funs], titles=[r'$r$',r'$g(r)$'], save = '../../InkscapeFigures/TSP_407.pdf')
```
